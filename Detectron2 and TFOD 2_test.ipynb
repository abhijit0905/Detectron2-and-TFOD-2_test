```python
#2
# Step 1: Install necessary libraries (run in the notebook if not installed)
# !pip install detectron2 torch torchvision
# !pip install requests matplotlib

import torch
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
from detectron2 import model_zoo
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg
from detectron2.utils.visualizer import Visualizer
from detectron2.data import MetadataCatalog

# Step 2: Load a pre-trained Detectron2 model (e.g., Mask R-CNN R50)
cfg = get_cfg()
# You can change the model to any pre-trained Detectron2 model available in the model zoo
cfg.merge_from_file(model_zoo.get_config_file("COCO-Detection/mask_rcnn_R_50_FPN_3x.yaml"))
cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-Detection/mask_rcnn_R_50_FPN_3x.yaml")
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5  # Set a confidence threshold (default is 0.5)

# Create a predictor object from the configuration
predictor = DefaultPredictor(cfg)

# Step 3: Load an image from a URL
image_url = 'https://example.com/your-image.jpg'  # Replace with the actual URL
response = requests.get(image_url)
image = Image.open(BytesIO(response.content))

# Step 4: Perform inference on the image
# Convert the PIL image to a numpy array (required by Detectron2)
image_np = np.array(image)

# Get the prediction results
outputs = predictor(image_np)

# Step 5: Visualize the results (bounding boxes, class labels)
v = Visualizer(image_np[:, :, ::-1], MetadataCatalog.get(cfg.DATASETS.TRAIN[0]), scale=1.2)
v = v.draw_instance_predictions(outputs["instances"].to("cpu"))

# Display the image with bounding boxes and labels
plt.imshow(v.get_image())
plt.axis('off')
plt.show()
```

```python
#4
# Install the required packages
!pip install tensorflow tensorflow-hub tensorflow-object-detection-api opencv-python matplotlib
import tensorflow as tf
import numpy as np
import cv2
import matplotlib.pyplot as plt
import requests
from io import BytesIO
from PIL import Image

# Step 1: Load a pre-trained model (e.g., SSD MobileNet V2)
MODEL_NAME = 'ssd_mobilenet_v2_coco'
MODEL_URL = f"http://download.tensorflow.org/models/object_detection/{MODEL_NAME}.tar.gz"

# Load the model from the TF model zoo
detect_fn = tf.saved_model.load(MODEL_URL)

# Step 2: Load the online image
image_url = 'https://your-image-url.com/sample-image.jpg'  # Replace with your image URL
image_response = requests.get(image_url)
image = Image.open(BytesIO(image_response.content))

# Step 3: Preprocess the image
image_np = np.array(image)

# Convert image to RGB (if it is in RGBA or other format)
if image_np.shape[-1] != 3:
    image_np = cv2.cvtColor(image_np, cv2.COLOR_RGBA2RGB)

# Convert to tensor and add batch dimension
input_tensor = tf.convert_to_tensor(image_np)
input_tensor = input_tensor[tf.newaxis,...]

# Step 4: Run inference on the image
# Perform object detection
output_dict = detect_fn(input_tensor)

# All outputs are numpy arrays, so convert to arrays if required
output_dict = {key:value.numpy() for key,value in output_dict.items()}

# Step 5: Visualize the results
# The output_dict contains detection boxes, scores, and classes
boxes = output_dict['detection_boxes']
classes = output_dict['detection_classes']
scores = output_dict['detection_scores']

# Set a confidence threshold to filter predictions
confidence_threshold = 0.5

# Filter out boxes with low confidence scores
valid_boxes = boxes[scores > confidence_threshold]
valid_classes = classes[scores > confidence_threshold]
valid_scores = scores[scores > confidence_threshold]

# Draw the bounding boxes and labels
image_with_boxes = image_np.copy()
for i in range(valid_boxes.shape[0]):
    ymin, xmin, ymax, xmax = valid_boxes[i]
    cv2.rectangle(image_with_boxes, 
                  (int(xmin * image_np.shape[1]), int(ymin * image_np.shape[0])), 
                  (int(xmax * image_np.shape[1]), int(ymax * image_np.shape[0])), 
                  (0, 255, 0), 2)  # Green color and thickness of 2
    
    label = f"Class: {int(valid_classes[i])}, Score: {valid_scores[i]:.2f}"
    cv2.putText(image_with_boxes, label, 
                (int(xmin * image_np.shape[1]), int(ymin * image_np.shape[0]) - 10), 
                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# Step 6: Display the image with bounding boxes
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(image_with_boxes, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Hide axes
plt.show()
```

```python
#6 
# Install dependencies
!pip install tensorflow tensorflow-object-detection-api pillow matplotlib

import tensorflow as tf
import numpy as np
import pathlib
import cv2
import requests
from io import BytesIO
from PIL import Image
from object_detection.utils import label_map_util
from object_detection.utils import visualization_utils as vis_util

# 1. Download and load the pre-trained model
MODEL_NAME = 'ssd_mobilenet_v2_coco'
MODEL_URL = f"http://download.tensorflow.org/models/object_detection/{MODEL_NAME}.tar.gz"
MODEL_DIR = pathlib.Path(f"./{MODEL_NAME}")

if not MODEL_DIR.exists():
    MODEL_DIR.mkdir()
    model_file = tf.keras.utils.get_file(f"{MODEL_NAME}.tar.gz", MODEL_URL, untar=True, cache_dir=str(MODEL_DIR))

PATH_TO_CKPT = MODEL_DIR / 'saved_model/saved_model.pb'
detect_fn = tf.saved_model.load(str(PATH_TO_CKPT))

# 2. Load the image
image_url = 'https://your-image-url.com/sample-image.jpg'
response = requests.get(image_url)
image = Image.open(BytesIO(response.content))
image_np = np.array(image)

input_tensor = tf.convert_to_tensor(image_np)
input_tensor = input_tensor[tf.newaxis,...]

# 3. Run inference
detections = detect_fn(input_tensor)
boxes = detections['detection_boxes'][0].numpy()
classes = detections['detection_classes'][0].numpy().astype(np.int32)
scores = detections['detection_scores'][0].numpy()

# 4. Load label map
LABEL_MAP_PATH = 'models/research/object_detection/data/mscoco_label_map.pbtxt'
label_map = label_map_util.load_labelmap(LABEL_MAP_PATH)
categories = label_map_util.convert_label_map_to_categories(label_map, max_num_classes=90, use_display_name=True)
category_index = label_map_util.create_category_index(categories)

# 5. Filter and visualize detections
threshold = 0.5
valid_indices = np.where(scores > threshold)
valid_boxes = boxes[valid_indices]
valid_classes = classes[valid_indices]
valid_scores = scores[valid_indices]

vis_util.visualize_boxes_and_labels_on_image_array(
    image_np,
    valid_boxes,
    valid_classes,
    valid_scores,
    category_index,
    instance_masks=None,
    use_normalized_coordinates=True,
    line_thickness=8,
    min_score_thresh=threshold
)

# 6. Display the image with bounding boxes and labels
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
plt.imshow(image_np)
plt.axis('off')
plt.show()
```
